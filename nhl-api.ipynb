#  NHL API Explorer

Designed to explore the capabilities of [nhl-api-py](https://pypi.org/project/nhl-api-py/)

```python
import pandas as pd
from nhlpy import NHLClient
```

```python
client = NHLClient()
```

## Teams

```python
# Get all teams
teams_json = client.teams.teams()
```

```python
teams_df = pd.DataFrame(teams_json).set_index('abbr') # make team abbr the index

# Extract conference and division abbr into new columns, replace original with names
teams_df['conference_abbr'] = teams_df['conference'].map(lambda v: v.get('abbr') )
teams_df['conference'] = teams_df['conference'].map(lambda v: v.get('name') )
teams_df['division_abbr'] = teams_df['division'].map(lambda v: v.get('abbr'))
teams_df['division'] = teams_df['division'].map(lambda v: v.get('name') )

teams_df
```
